```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Carga de el dataFrame
df = pd.read_csv('../data/vgsales.csv')
## Leer el dataFrame
df.head()
```

```python
## Uso de la momoria
df.memory_usage()
```

```python
df[['Name', 'Year']].head(30)
```

```python
##Filtramos por nombre en toda la data frame

filtro = df['Name'].str.contains('Grand Theft Auto', case =False, na=False)

df.loc[filtro, ['Name','Year','Publisher','Platform','NA_Sales']].head()
```

```python
df_filtrado = df.loc[filtro, ['Name','Year','Publisher','Platform','NA_Sales']].head()
df_filtrado.to_csv('datosGrand_Theft_Auto.csv', index=False)
```

```python
df_filtrado2 = df.loc[filtro, ['Name','Year']].head()
df_filtrado2_ordenado = df_filtrado2.sort_values(by='Year')#Filtrado de menor a mayor
df_grafico = df_filtrado2_ordenado.head()#Creamos uno nuevo con los datos acomodados

# Creamos el grafico
plt.figure(figsize=(15,8))
plt.bar(df_grafico['Name'],df_grafico['Year'] , color='blue')
plt.xlabel('Juego')
plt.ylabel('Ao de lanzamiento')
plt.title('Ao de lanzamiento de GTA')
plt.ylim(2000, 2015)#Limite de rango de fechas
plt.show()
```
